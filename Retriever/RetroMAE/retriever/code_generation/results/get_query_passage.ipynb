```python
with open("../results_nl_all/results_{}/query_passage.json".format(type_file),"r") as f:
    with open("../result_nl_all/results_{}/query_many_passage.json".format(type_file),"w") as f1:
        line = eval(f.readline())
        flag = line['query_id']
        many_passage = [(line['passage'],line['get_code'])]
        new_line = {}
        new_line['id'] = flag
        new_line['query'] = (line['query'],line['real_code'])
        while True:
            temp = f.readline()
            if not temp:
                break
            line = eval(temp)
            if flag == line['query_id']:
                many_passage.append((line['passage'],line['get_code']))
            else:
                new_line['retrieved_passage'] = many_passage
                f1.write(json.dumps(new_line))
                f1.write('\n')
                flag = line['query_id']
                many_passage = [(line['passage'],line['get_code'])]
                new_line = {}
                new_line['id'] = flag
                new_line['query'] = (line['query'],line['real_code'])          
```

```python
# 最终得到的是nl cat 5 code
with open("../results_nl_all/results_{}/query_passage.json".format(type_file),"r") as f:
    with open("../results_nl_all/results_{}/retromae_{}.json".format(type_file,type_file),"w") as f1:
        # 刚开始读入第一行
        line = eval(f.readline())
        # i作为标识符控制写入文件是第几个
        i = 1
        count = 0
        # flag决定当前query是否读完以及是否要讲查询结果写入文件中
        flag = line['query_id']
        # 向nl中添加检索到的代码 retrieved_code作为标识符
        query_code = " retrieved_code "
        query_code += line['get_code']
        new_line = {}
        # 只保留最开始的nl
        new_line['nl'] = line['query']
        
        while True:
            i += 1
            temp = f.readline()
            # 读到文件结尾 退出程序
            if not temp:
                print(count)
                count+=1
                new_line['nl'] += query_code
                f1.write(json.dumps(new_line))
                f1.write('\n')
                flag = line['query_id']
                query_code = " retrieved_code "
                query_code += line['get_code']
                new_line = {}
                new_line['nl'] = line['query']
                break
            line = eval(temp)
            if i <= 5 and flag == line['query_id']:
                query_code += " retrieved_code "
                query_code += line['get_code']
                if i==5:
                    new_line['code'] = line['real_code']         
            else:
                if flag == line['query_id']:
                    continue
                else:
                    count+=1
                    new_line['nl'] += query_code
                    f1.write(json.dumps(new_line))
                    f1.write('\n')
                    flag = line['query_id']
                    query_code = " retrieved_code "
                    query_code += line['get_code']
                    new_line = {}
                    new_line['nl'] = line['query']
                    i = 1
```
